## (d)Machine Learning Models
#### Cluster data 
#### Classify data and perform LR model and DT model for predictions

```python
#Import pandas, seaborn, matplotlib
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

#read data 
cust_behave = pd.read_csv('data/Customer_Behaviour.csv')
cust_char = pd.read_csv('data/Customer_Profiles.csv')
reportA = pd.merge(cust_char, cust_behave, on = 'CustomerId' )
```

```python
# Select the columns
reportA = reportA[['NumWebPurchases','Income','NumWebVisitsMonth','NumStorePurchases']]
```

```python
#Normalize the data by subtracting the column means and dividing by col stdv
reportA = (reportA - reportA.mean())/reportA.std()
# Cluster with K-Means
mykmeans = KMeans(n_clusters=3).fit(reportA)
mykmeans.labels_
# Evaluate clusters using PCA 
mypca = PCA(n_components=2)
myscores = mypca.fit_transform(reportA)

myscores = pd.DataFrame(myscores, columns=["PC1", "PC2"])
sns.scatterplot(data=myscores,
                x="PC1", 
                y="PC2", 
                c=mykmeans.labels_)
#Describe the clusters 
myloadings = pd.DataFrame(mypca.components_.T, 
             index=reportA.columns,
             columns=["PC1","PC2"])
myloadings
```

### Time to perform machine learning models

```python
# Import pandas, seaborn, and matplotlib.
import pandas as pd
from sklearn import metrics
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier

#Read train data
df1 = pd.read_csv('data/Customer_behaviour_train.csv')
df2 = pd.read_csv('data/Customer_Profiles_train.csv')
train_data = pd.merge(df2, df1, on = 'CustomerId' )

#Read test data
df13 = pd.read_csv('data/Customer_behaviour_test.csv')
df23 = pd.read_csv('data/Customer_Profiles_test.csv')
test_data = pd.merge(df23, df13, on = 'CustomerId' )
```

```python
predictors = ['NumWebPurchases','Income','Kidhome','Teenhome','NumWebVisitsMonth']
X_train = pd.get_dummies(train_data[predictors],
               drop_first=True)
X_test = pd.get_dummies(test_data[predictors],
               drop_first=True)

y_train = train_data['NumStorePurchases']
```

```python
# Fit logistic regression model and calculate accuracy using the train data.
my_lr = LogisticRegression(max_iter=400).fit(X_train, y_train)
lr_pred_train = my_lr.predict(X_train)
metrics.accuracy_score(y_train, lr_pred_train)
```

```python
#Create confusion matrix
metrics.confusion_matrix(y_train, lr_pred_train)
```

```python
#Make predictions using the data 
lr_pred_test = my_lr.predict(X_test)
```

```python
lr_output = pd.DataFrame(lr_pred_test,
                      index=X_test.index,
                      columns=["NumStorePurchasesp"])
lr_output.to_csv("data/lr_pred.csv")
```

```python
# Visualize the accuracy of LR model
import matplotlib.pyplot as plt

lr_pred = pd.read_csv("data/lr_pred.csv")
actual = pd.read_csv('data/Customer_Behaviour_test.csv')

merged = pd.merge(lr_pred, actual, on = 'CustomerId')

plt.plot(merged[['CustomerId']][:30], merged[['NumStorePurchases']][:30], label = "Actual")
plt.plot(merged[['CustomerId']][:30], merged[['NumStorePurchasesp']][:30], label = "Predicted")
plt.legend()
plt.show()
```

```python
#Fit decision tree and calculate accuracy on train data 
my_tree = DecisionTreeClassifier().fit(X_train, y_train)
tree_pred_train = my_tree.predict(X_train)
metrics.accuracy_score(y_train, tree_pred_train)
```

```python
#Create confusion matrix for train data
pd.DataFrame(metrics.confusion_matrix(y_train, tree_pred_train, normalize="true"))
```

```python
#Make predictions
tree_pred_test = my_tree.predict(X_test)
```

```python
#Write to CSV
tree_output = pd.DataFrame(tree_pred_test,
                      index=X_test.index,
                      columns=["NumStorePurchasesp"])
tree_output.to_csv("data/tree_pred.csv")
```

```python
# Visualize the accuracy of DT model
import matplotlib.pyplot as plt

lr_pred = pd.read_csv("data/tree_pred.csv")
actual = pd.read_csv('data/Customer_Behaviour_test.csv')

merged = pd.merge(lr_pred, actual, on = 'CustomerId')

plt.plot(merged[['CustomerId']][:30], merged[['NumStorePurchases']][:30], label = "Actual")
plt.plot(merged[['CustomerId']][:30], merged[['NumStorePurchasesp']][:30], label = "Predicted")
plt.legend()
plt.show()
```
