# 1 - Analysis of customer profile to customer behavior
### More particularly we want to analyze customer characteristics vs web purchases, deal purchases, deal purchases


```python
# Import customer behavior table
# Import customer characteristics table 
cust_behave = pd.read_csv('data/Customer_Behaviour.csv')
cust_char = pd.read_csv('data/Customer_Profiles.csv')
```

## (a) In order to perform analysis we need to merge the data
#### We used Inner join because both tables have CustomerID in common which allows the data to merge without null values

```python
report1 = pd.merge(cust_char, cust_behave, on = 'CustomerId' )
```

## (b) Lets do some simple statistical calculations to understand our data
#### Find avg customer profile based grouped by web purchases
#### Find avg customer profile based grouped by store purchases

```python
report1.groupby(by = ['NumWebPurchases']).mean()
```

```python
report1.groupby(by = ['NumStorePurchases']).mean()
```
